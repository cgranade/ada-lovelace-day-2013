# Fractals

## A Brief Diversion Into Complex Numbers

The square root $\sqrt{z}$ of a number $z$ is often introduced as the number $y$ such that $y^2 = z$. Using this definition, we can readily figure out the square roots of positive numbers. For example, knowing that $2^2 = 4$, we also know that $\sqrt{4} = 2$.

What about $\sqrt{-1}$, though? What number could we possibly square to get $-1$? Neither positive numbers nor negative numbers can do it, so if we want the square root to be defined for negative numbers as well, then we need some other *kind* of number. Since mathematicians love giving things names, let's call $\sqrt{-1} = i$, for *imaginary*. Then, by definition, $i$ is the number such that $i^2 = -1$. This is a very cool number indeed, as it lets us start playing around not just with the real numbers, but the *complex* numbers, formed by adding a real number to a real number times $i$, such as $a + bi$.

Addition of complex numbers follow the normal rules of algebra. For instance,
$(a + bi) + (c + di) = (a + c) + (b + d)i$. Multiplication is almost as simple, with the additional rule that $i^2 = -1$:
$(a + bi) \times (c + di) = a(c+di) + bi(c + di) = ac + adi + bci - bd = (ac - bd) + (ad + bc)i$.

Python has support for complex numbers built-in, using the function `complex`. The biggest thing to watch out for is that in Python, $i$ is spelled `j`.

```python
z = complex(1, 2)
print z
```

```python
print z * z, z + z
```

```python
print z + z == 2 * z
```

In the same way we can ask how far a point $(x, y)$ is from the origin, we can ask how far a complex number $a + bi$ would be from the origin if we plotted the real part as $x$ and the complex part as $y$. This is called the *modulus* of a complex number (written $\|z\|$), and is calculated in Python using the `abs` function:

```python
print abs(z), abs(z**2)
```

## Julia Set

Given a function of the form $f(z) = z^2 + c$ where $z$ and $c$ are allowed to be complex numbers, we can ask what happens if we iteratively apply $f$ to some number $z$, as in $f(f(\cdots f(z)\cdots))$. Roughly, for a given function $f$, the [Julia set](http://en.wikipedia.org/wiki/Julia_set) $J(f)$ is the set of points that behave *chaotically* as $f$ is applied over and over again. 

Let $f^n(z)$ mean "apply $f$ $n$ times to $z$". Then, an interesting way of visualizing the structure of $J(f)$ is to plot for each $z$ the smallest $n$ such that $\|f^n(z)\| > E$ for some number $E$.

To implement this in Python, then, we need two things:

- A way of repeatedly applying $f$ to some number $z$.
- A way of keeping track of when $\|f^n(z)\|$ "escapes".

We can do this by creating a function that takes in a function we want to iterate,
a complex number $z_0$ we will start the iteration at, a number of iterations we're
willing to compute before giving up, and the number $E$ we use to decide whether a
point has "escaped".

```python
def julia_color(fn, z0, max_iter=4500, escape=10000.0):
    # f^0(z) = z
    z = z0
    
    for idx_iter in xrange(max_iter):
        # Apply fn to z and save the result back to z.
        z = fn(z)
        
        # Has z escaped yet? If so, return how long it took.
        if abs(z) > escape:
            return idx_iter
        
    # If we ran out of iterations, just return the maximum number
    # of iterations we were willing to run.
    return idx_iter
```

Next, we need to actually make a function to iterate over. The `lambda` keyword in Python is great for making short one-line functions, such as this one that creates a function $f(z) = z^2 - 0.4+0.6i$ (an especially pretty Julia fractal suggested by [Google Labs' fractal map](http://juliamap.googlelabs.com/)).

```python
j_fn = lambda z: z**2 + (-0.4+0.6j)
```

Now we can draw it, using the `imshow` plotting function, which will automatically return all of the iteration numbers we made above into pretty colors. To make a map of colors for `imshow`, we use an `array` to hold the number of iterations each complex number $x + yi$ takes to escape. The `empty` function makes a new array with nothing in it, but of a given size (in this case, 200 by 200 works well). Finally, the `linspace` function takes a range and a number of points, and returns an array containing numbers evenly spread out over that range.

```python
n_x = 600
n_y = 600

colors = empty((n_x, n_y))
xs = linspace(-1, 1, n_x)
ys = linspace(-1, 1, n_y)

for idx_x in xrange(n_x):
    for idx_y in xrange(n_y):
        colors[idx_x, idx_y] = julia_color(j_fn, complex(xs[idx_x], ys[idx_y]))
        
figure(figsize=(n_x / 80, n_y / 80))
imshow(colors, extent=(-1, 1, -1, 1));
```
